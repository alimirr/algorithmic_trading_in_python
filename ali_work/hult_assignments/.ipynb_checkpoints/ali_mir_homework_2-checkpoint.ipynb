1. Was given a simple moving average (SMA) strategy of 50 and 200 days respectively.
1. Was tasked to find strategy, in which the SMA days perform better than the one described above
1. Used a loop to assess various nubmer of days and chose **35 for short** and **350 for long**.

# Description of Strategies
Strategy used was a trend following strategy. **Trend following**, also known as a momentum strategy, **bets that the price trend will continue in the same direction**. When the price rises above its moving average, enter a long position to bet the price will continue to rise. 

* **long signal**: When the short-term SMA crosses above the long-term SMA, as it suggests the price is picking up momentum. 

* **short signal**: When the short-term SMA crosses below the long-term SMA, as it suggests that the price is losing momentum.

We use these signals to trigger a buy/sell response. 

# Constructing the Default Strategy
To construct the default strategy which has a short signal of 50 days and and a long signal of 200 days. We first construct the rolling average of the four tickers. Then we define the strategy using the bt package. The condition of the trend following is in the **bt.algos.SelectWhere()** function where the condition is when when price > SMA. 

We then backtest the strategy against historical data and plot the resulting index.

# Costructing a Better Strategy
In order to construct a better strategy in a specific range, a loop was constructed with varying number.

* **Short Signal**: 10 different number of days were test ranging from 20 to 70 with intervals of 5. 
* **Long Signal**: 24 different number of days were test ranging from 150 to 390 with intervals of 10.
    
Each Signal was backtest and plotted against the default strategies and the best among them were chosen.

# Results
For the short and long signal, the best number of days for the SMA was 35 and 350, respectively. While the better strategies did perform well overall, there were times where they versed worse against the default strategy as can be seen in the figure below. Both 35 SMA and 350 SMA fared better than the default strategy overall, garnering return of ~178% and ~205%, much higher than the ~137% and ~117%, respectively. 

According to the metrics displayed in the table below, the better strategy fares better in most aspects. The total return is higher for the better strategy and so is the cagr, daily and monthly Sharpe Ratio, average drawdown and month/yearly mean. Interestingly, however, the yearly Sharpe ratio of the default above 50 SMA is higher than all other SMAs. This implies that on a yearly basis, above 50 SMA may be a safer investment. However, assessing from the figure below, it could be because the above 50 SMA is more sensitive stock falls and less sensitive to price climbs. The types of tickers should be gauged properly before using strategies. For example, in this case, the 4 tickers are FAANG companies that fared well in the past few years and have mostly seen a rise in stock value.

The **total return** of the above 35 SMA strategy is 110% more than the above 50 SMA strategy. Similarly, the above 350 strategy is a whopping ~517% increase from the default above 200 SMA strategy.

```python
import bt
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import os
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
# getting ticker data
price_data = bt.get('fb, aapl, amzn, nflx', start='2018-1-1', end='2021-6-1')
```

```python
# setting up SMAs
short = 35
long = 350

short_sma_df = price_data.rolling(short).mean()
long_sma_df = price_data.rolling(long).mean()

sma_50_df = price_data.rolling(50).mean()
sma_200_df = price_data.rolling(200).mean()
```

```python
# Backtesting and acquiring reuslts of chosen strategies

bt_short_strategy = bt.Strategy(f'my_above_{short}_SMA',
                                [bt.algos.SelectWhere(price_data > short_sma_df),
                                 bt.algos.WeighEqually(),
                                 bt.algos.Rebalance()])

bt_short_backtest = bt.Backtest(bt_short_strategy, price_data)
bt_short_result = bt.run(bt_short_backtest)

bt_long_strategy = bt.Strategy(f'my_above_{long}_SMA',
                               [bt.algos.SelectWhere(price_data > long_sma_df),
                                bt.algos.WeighEqually(),
                                bt.algos.Rebalance()])

bt_long_backtest = bt.Backtest(bt_long_strategy, price_data)
bt_long_result = bt.run(bt_long_backtest)

bt_50_strategy = bt.Strategy('Above_50_SMA',
                                [bt.algos.SelectWhere(price_data > sma_50_df),
                                 bt.algos.WeighEqually(),
                                 bt.algos.Rebalance()])

bt_50_backtest = bt.Backtest(bt_50_strategy, price_data)
bt_50_result = bt.run(bt_50_backtest)

bt_200_strategy = bt.Strategy('Above_200_SMA',
                               [bt.algos.SelectWhere(price_data > sma_200_df),
                                bt.algos.WeighEqually(),
                                bt.algos.Rebalance()])

bt_200_backtest = bt.Backtest(bt_200_strategy, price_data)
bt_200_result = bt.run(bt_200_backtest)

my_strategy_df = bt.merge(bt_long_result.prices, bt_short_result.prices, bt_200_result.prices, bt_50_result.prices)
```

```python
# Combining the results of all different SMAs into a dataframe
display_df = bt_50_result.lookback_returns.append(bt_50_result.stats)
display_df = display_df.join(bt_200_result.lookback_returns.append(bt_200_result.stats))
display_df = display_df.join(bt_short_result.lookback_returns.append(bt_short_result.stats))
display_df = display_df.join(bt_long_result.lookback_returns.append(bt_long_result.stats))

display_df = display_df.drop_duplicates()
display_df.to_csv('./assets/display_table.csv')
display_df
```

```python
# Plotting graphs of the different SMAs

my_strategy_df.plot(title='My SMA Backtests vs 200 and 50 SMA', # plotting backtest output
                    figsize=(15,6), 
                    color=['c','orange','grey','pink'])

plt.axhline(y=my_strategy_df.iloc[-1,0], color='c', linestyle='--') # Horizontal line
plt.text('2021-8-7',210,round(my_strategy_df.iloc[-1,0], 2), color='c') # Horizontal line text

plt.axhline(y=my_strategy_df.iloc[-1,1], color='orange', linestyle='--') 
plt.text('2021-8-7',round(my_strategy_df.iloc[-1,1], 2),round(my_strategy_df.iloc[-1,1], 2), color='orange')

plt.axhline(y=my_strategy_df.iloc[-1,2], color='grey', linestyle='--')
plt.text('2021-8-7',round(my_strategy_df.iloc[-1,2],2),round(my_strategy_df.iloc[-1,2], 2), color='grey')

plt.axhline(y=my_strategy_df.iloc[-1,3], color='pink', linestyle='--')
plt.text('2021-8-7',round(my_strategy_df.iloc[-1,3], 2),round(my_strategy_df.iloc[-1,3], 2), color='pink')

plt.ylabel('Index')
plt.xlabel('Date')

# Uncomment below 2 lines to save plot in assets folder in root
# os.makedirs('./assets', exist_ok=True)
# plt.savefig('./assets/hw_2_1')
plt.show()
```
